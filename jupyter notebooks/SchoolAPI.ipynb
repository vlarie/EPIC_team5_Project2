```python
#receive lat and lng (x, y)
#identify which school district
import requests
import json
import pandas as pd
import numpy as np
import xml.etree.ElementTree as ET


URL = 'https://api.greatschools.org/schools/nearby'
```

```python
# masterData_csv = pd.read_csv("./Data/latlon_schools_NA.csv")
masterData_csv = pd.read_csv("./Data/masterDataCLEAN.csv")
master_df = pd.DataFrame(masterData_csv)
```

```python
# masterData_csv = pd.read_csv("./Data/latlon_schools_NA.csv")
masterData_csv = pd.read_csv("./Data/masterDataCLEAN.csv")
master_df = pd.DataFrame(masterData_csv)

# Loop through Requests (Partial Params):
alat = []
alon = []
s_name = []
zip_code = []
ratings = []

counter = 1

#schooltypes = ["elementary-schools", "middle-schools", "high-schools"]

for index, row in master_df.tail(1000).iterrows():
    params = {
    'key' : '41c6b7482243181cbb1f77b88b04eb12',
    'state' : 'TX',
    'lat' : row['alat'],
    'lon' : row['alon'], 
    'limit' : 1,
    'schoolType' : 'public',       
    'levelCode' : 'high-schools',
    'radius' : 25    
    }

    school_list = requests.get(URL, params=params)
#    print(school_list)
    try:
        if school_list.status_code == 200:
            new_school = school_list.text
            root = ET.fromstring(new_school)
    #school_list.text

            alat.append(row['alat'])
            alon.append(row['alon'])

            for child in root:
                print(child[1].text, child[5].text)
                s_name.append(child[1].text)
                zip_code.append(child[12].text)
                
                rating = child[5].text
                if rating.isnumeric():
                    ratings.append(rating)
                else:
                    ratings.append(np.NAN)

                print(f"Processed Row # {counter}.")
                counter += 1
    except:
        print("Error!!")
        pass

    
print("Finished Requests.")
```
