```python
import pandas as pd
import numpy as np
```

```python

filenames = ["Elementary_School_Ratings.csv", "middle_School_Ratings.csv", "High_School_Ratings.csv", "new_High_School_Ratings.csv"] # Fill in remaining files.
df = pd.DataFrame()
for filename in filenames:
    print(filename)
    df = df.append(pd.read_csv(filename, encoding="ISO-8859-1"), sort=False)
```

```python
df
```

```python
df["zipcode"] = ""
```

```python
for index, row in df.iterrows():
    try: 
        zipcode = (row["School Address"].split("TX  ")[1])
        df.set_value(index, "zipcode", zipcode)
    except:
        df.set_value(index, "zipcode", np.nan)
#     print(row["School Name"])
    if ("Jail" in row["School Name"]):
        df.drop(index)
        print("found jail")
    if ("Hospital" in row["School Name"]):
        df.drop(index)
    
    
```

```python
df = df.dropna()
df.drop_duplicates(subset=None, keep='first', inplace=False)
```
