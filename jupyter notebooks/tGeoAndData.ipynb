```python
#  import old master data

fileImport = pd.read_csv('../data/zillowCommuteData.csv')
masterDF = pd.DataFrame(fileImport)
```

```python
masterDF.head()
```

```python
masterByZipCount = masterDF.groupby(by="zipcode").size()

masterByZipCountDF = pd.DataFrame({'zipcode':masterByZipCount.index, 'count':masterByZipCount.values})

masterByZipCountDF
```

```python

masterByZipDF = masterDF.groupby(by="zipcode").mean()
```

```python
masterByZipDF.head()
```

```python
masterByZipDF = masterByZipDF.drop(['zid','alat','alon','last_sold_price'], axis=1)
masterByZipDF.head()
```

```python
masterByZipDF.reset_index(level=0, inplace=True)
masterByZipDF.head()
```

```python
masterByZipDF = pd.merge(masterByZipDF,masterByZipCountDF[['zipcode','count']],on='zipcode', how='left')
masterByZipDF.head()
```

```python
masterByZipDF.zipcode = masterByZipDF.zipcode.astype(str)
masterByZipDF.valuation = masterByZipDF.valuation.astype(int)
masterByZipDF.sqft = masterByZipDF.sqft.astype(int)
masterByZipDF.lot_sqft = masterByZipDF.lot_sqft.astype(int)
masterByZipDF.year_built = masterByZipDF.year_built.astype(int)
masterByZipDF.commuteTime = masterByZipDF.commuteTime.astype(int)
masterByZipDF.dtypes
```

```python
fileImport = pd.read_csv('../data/crimeData.csv')
crimeDF = pd.DataFrame(fileImport)
```

```python
crimeDF.head()
```

```python
geoWithEpicProps = geoDF.merge(masterByZipDF, on='zipcode')
print(len(geoWithEpicProps))
geoWithEpicProps.head()
```

```python
zipsToDrop = [78620,78669,78619]
newDF = gpd.GeoDataFrame()

for index, row in geoWithEpicProps.iterrows():
    keep = True
    for zip in zipsToDrop:
        if int(row['zipcode']) == zip:
            print(row['zipcode'] + " " + str(zip))
            keep = False
    if keep:
        newDF = newDF.append(geoWithEpicProps.iloc[index])
        print(index)
len(newDF)
```

```python
newDF.to_file('../data/EPIC_data_noCrime.geojson', driver="GeoJSON")
```
